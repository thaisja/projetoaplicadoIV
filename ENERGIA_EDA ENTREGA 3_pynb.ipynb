## **Referêncial Teórico**

A análise de séries temporais tem sido amplamente utilizada em problemas
relacionados à previsão de demanda energética, constituindo uma ferramenta
essencial para o planejamento e a otimização de recursos. Diversos estudos
demonstram a eficácia tanto de modelos estatísticos quanto de técnicas de
aprendizado de máquina na modelagem de dados temporais.

Entre as abordagens estatísticas clássicas, destacam-se os modelos ARIMA
(AutoRegressive Integrated Moving Average), amplamente empregados para
capturar padrões de dependência temporal em séries estacionárias. Esses modelos
apresentam como principais vantagens a interpretabilidade e a robustez em
cenários com dados estruturados, embora possam apresentar limitações na
modelagem de padrões não lineares (HYNDMAN; ATHANASOPOULOS, 2021).

Para séries que apresentam sazonalidade, como é o caso do consumo
energético, os modelos SARIMA (Seasonal ARIMA) são frequentemente utilizados,
permitindo incorporar padrões periódicos na modelagem. No entanto, esses
modelos exigem maior cuidado na definição de parâmetros e pressupõem certa
regularidade nos padrões sazonais. Além dos métodos estatísticos, técnicas de
aprendizado de máquina, como redes neurais e modelos baseados em árvores, têm
sido aplicadas na previsão de consumo energético. Essas abordagens apresentam
bom desempenho na captura de relações não lineares e interações complexas entre
variáveis. Por outro lado, podem demandar maior volume de dados e apresentam
menor interpretabilidade em comparação aos modelos tradicionais.

De modo geral, a escolha da técnica mais adequada depende das
características da série temporal analisada, como a presença de tendência,
sazonalidade e nível de ruído. No contexto deste trabalho, a identificação dessas
componentes por meio da análise exploratória fundamenta a escolha de modelos
que considerem dependência temporal e padrões sazonais, como os modelos
ARIMA e SARIMA.

4. ANÁLISE DE AUTOCORRELAÇÃO (ACF e PACF)
A autocorrelação é uma medida estatística que permite avaliar a relação entre valores passados e presentes de uma série temporal, sendo fundamental para identificar padrões de dependência temporal (HYNDMAN; ATHANASOPOULOS, 2021).
A análise da função de autocorrelação (ACF) evidencia a existência de dependência temporal na série, com valores significativos nos primeiros lags. Esse comportamento indica que os valores atuais do consumo energético estão relacionados a observações anteriores.
Observa-se ainda que, à medida que o número de defasagens aumenta, os coeficientes de autocorrelação tendem a se aproximar de zero, sugerindo redução da influência de valores mais distantes no tempo.
Por sua vez, a função de autocorrelação parcial (PACF) apresenta poucos lags com relevância estatística, indicando que a estrutura da série pode ser explicada por um número limitado de termos autoregressivos.
Em conjunto, esses resultados indicam que a série possui dependência temporal significativa e é adequada para modelagem por meio de técnicas estatísticas de séries temporais.
5. DECOMPOSIÇÃO DA SÉRIE TEMPORAL
A decomposição da série temporal permite identificar seus principais componentes estruturais — tendência, sazonalidade e resíduo — sendo amplamente utilizada na análise de séries temporais (HYNDMAN; ATHANASOPOULOS, 2021).
A componente de tendência evidencia variações ao longo do período analisado, com níveis mais elevados no início do ano, redução ao longo dos meses intermediários e posterior recuperação no final do período. Esse comportamento indica a influência de fatores sazonais de longo prazo no consumo energético.
A componente sazonal apresenta um padrão altamente regular e repetitivo, evidenciando forte sazonalidade associada ao ciclo diário de consumo energético.
Por fim, o componente residual não apresenta estrutura definida, caracterizando-se como ruído aleatório, indicando que as principais variações da série foram capturadas pelas componentes de tendência e sazonalidade.
6. IDENTIFICAÇÃO DAS COMPONENTES DA SÉRIE

Tendência não linear ao longo do tempo

**Modelo Base**


Nesta etapa, foi desenvolvido um modelo base para previsão do consumo de energia elétrica, utilizando o modelo ARIMA (AutoRegressive Integrated Moving Average). A escolha desse modelo foi fundamentada nas análises realizadas anteriormente, especialmente nas funções de autocorrelação (ACF) e autocorrelação parcial (PACF), que indicaram a presença de dependência temporal e a adequação de modelos autoregressivos.

Inicialmente, a série temporal foi dividida em conjuntos de treino e teste, utilizando 80% dos dados para treinamento e 20% para validação. Em seguida, foi ajustado um modelo ARIMA com parâmetros (2,1,2), considerando a necessidade de diferenciação para tratar a não estacionariedade da série.

Após o treinamento, foram geradas previsões para o conjunto de teste, permitindo comparar os valores previstos com os valores reais observados.





```python
# ======================================
# MODELO BASE - ARIMA
# ======================================

from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# usar série diária
serie = serie_diaria

# divisão treino/teste (80% treino)
train_size = int(len(serie) * 0.8)
train, test = serie[:train_size], serie[train_size:]

# treinar modelo ARIMA
modelo = ARIMA(train, order=(2,1,2))
modelo_fit = modelo.fit()

# previsão
forecast = modelo_fit.forecast(steps=len(test))

# métricas
mae = mean_absolute_error(test, forecast)
mse = mean_squared_error(test, forecast)
rmse = np.sqrt(mse)

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")

# gráfico
plt.figure(figsize=(12,5))
plt.plot(train, label='Treino')
plt.plot(test, label='Real')
plt.plot(test.index, forecast, label='Previsto')
plt.legend()
plt.title('Modelo ARIMA - Previsão de Consumo de Energia')
plt.xlabel('Tempo')
plt.ylabel('Carga (MW)')
plt.show()
```

O desempenho do modelo foi avaliado por meio das métricas MAE (Mean Absolute Error) e RMSE (Root Mean Squared Error), que medem o erro médio das previsões.

Os resultados indicam que o modelo ARIMA é capaz de capturar parte do comportamento da série temporal, apresentando desempenho satisfatório como modelo inicial. No entanto, observa-se que o modelo apresenta limitações na captura de variações mais abruptas e padrões sazonais complexos, sugerindo a possibilidade de melhoria com modelos mais avançados, como SARIMA ou técnicas de aprendizado de máquina.

```python
import matplotlib.pyplot as plt

# Agrupar por subsistema
subsistemas = df['nom_subsistema'].unique()

plt.figure(figsize=(12,6))

for sub in subsistemas:
    serie_sub = df[df['nom_subsistema'] == sub]['val_cargaenergiamwmed']
    serie_sub = serie_sub.resample('D').mean()
    plt.plot(serie_sub, label=sub)

plt.legend()
plt.title('Consumo Médio Diário por Subsistema')
plt.show()
```

**ANÁLISE POR SUBSISTEMA**


A análise segmentada por subsistema evidencia diferenças significativas no comportamento do consumo energético entre as regiões do Brasil. Observa-se que o subsistema Sudeste/Centro-Oeste apresenta os maiores níveis de consumo ao longo de todo o período analisado, refletindo sua elevada concentração populacional e atividade econômica.

O subsistema Sul apresenta comportamento semelhante, porém em menor escala, enquanto o Nordeste apresenta níveis intermediários e maior estabilidade ao longo do tempo. Já o Norte apresenta os menores níveis de consumo, com menor variação relativa.

Além disso, todos os subsistemas apresentam padrões recorrentes ao longo do tempo, evidenciando a presença de sazonalidade, possivelmente associada aos ciclos diários de consumo energético. Também é possível observar variações ao longo do ano, indicando a presença de tendência e influência de fatores sazonais de maior escala.

Essa análise reforça a importância de considerar características regionais na modelagem da série temporal, uma vez que o comportamento do consumo energético varia significativamente entre os subsistemas.
